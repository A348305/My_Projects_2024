```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data loading and processing
df = pd.read_csv("https://raw.githubusercontent.com/A348305/Data-Science-Internship-Intern-Career/master/youtubers_df.csv")

# Display ther first 5 rows
df.head()
```

```python
# 1
# Display summery statistics
df.describe()
```

```python
# checking missing values
df.isnull().sum()
```

```python
#Checking for outliers
sns.boxplot(data=df[['Categories','Suscribers']])
plt.show()
```

```python
#Dispay correlation matrix
df.corr()
```

```python
# 2
# Count occurrences of each category
category_counts = df['Categories'].value_counts()
```

```python
# Plot the top catagories
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('Top YouTube Streaner Category')
plt.xlabel('Categories')
plt.ylabel('Number of Streamers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plot the correlation matrix
correlation_matrix = df[['Suscribers', 'Likes', 'Comments']].corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 3
# Count the accurrences of each country
country_counts = df['Country'].value_counts()
```

```python
# Plot the top countries
plt.figure(figsize=(12, 8))
sns.barplot(x=country_counts.index[:10], y=country_counts.values[:10], palette='muted')
plt.title('Top 10 Streamer Audiences by Country')
plt.xlabel('Country')
plt.ylabel('Number of Streamers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Group by Country and Category, then count tne number of streamers in each group
country_category_counts = df.groupby(['Country', 'Categories']).size().unstack(fill_value=0)


# plot a heapmap to visualise the distribution
plt.figure(figsize=(16, 10))
sns.heatmap(country_category_counts, cmap='Blues', annot=True, fmt='g', cbar_kws={'label': 'Number of Streamers'})
plt.title('Audience Distribution by Country and Category')
plt.xlabel('Content Category')
plt.ylabel('Country')
plt.show()
```

```python
# 4
# Calculate average matrics
average_metrics = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()


# DIsplay the average metrics
print("Average Subscribers:", average_metrics['Suscribers'])
print("Average Visits:", average_metrics['Visits'])
print("Average Likes:", average_metrics['Likes'])
print("Average Comments:", average_metrics['Comments'])
```

```python
# Visualizing the Average Matrics:
# Plot bar chart for average metrics
plt.figure(figsize=(10, 9))
sns.barplot(x=average_metrics.index, y=average_metrics.values, palette='pastel')
plt.title('Average Performance Matrics')
plt.xlabel('Metrics')
plt.ylabel('Average Value')
plt.show()
```

```python
average_metrics_by_Categories = df.groupby('Categories').mean()

# Plot bar chart for average metrics by Categories
plt.figure(figsize=(12, 8))
average_metrics_by_Categories.plot(kind='bar', colormap='viridis')
plt.title('Average Performance Metrics by Categories')
plt.xlabel('Categories')
plt.ylabel('Average Value')
plt.show()
```

```python
# 5
# Count the occurrences of each Categories
Categories_counts = df['Categories'].value_counts()

# Plot the distribution of content categories
plt.figure(figsize=(12, 8))
sns.barplot(x=Categories_counts.index, y=Categories_counts.values, palette='viridis')
plt.title('Distribution of Content Category')
plt.xlabel('Categories')
plt.ylabel('Number of Streamers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Calculate average metrics by 'Category'
average_metrics_by_Categories = df.groupby('Categories')[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()

# Display the average metrics by Categories
print(average_metrics_by_Categories)


# Plot bar chart for average metrics by category
plt.figure(figsize=(12, 8))
average_metrics_by_Categories.plot(kind='bar', colormap='viridis')
plt.title('Average Performance Metrics by Categories')
plt.xlabel('Categories')
plt.ylabel('Average Value')
plt.show()
```

```python
# 6
# Identify the top 50 users with the highest Suscribers, Visits, and Likes
top_suscribers_users = df.nlargest(50, 'Suscribers')
top_visits_users = df.nlargest(50, 'Visits')
top_likes_users = df.nlargest(50, 'Likes')

# Create a bar graph
plt.figure(figsize=(12, 6))
sns.barplot(x='Username', y='Suscribers', data=top_suscribers_users, color='blue', label='Suscribers')
sns.barplot(x='Username', y='Visits', data=top_visits_users, color='orange', label='Visits')
sns.barplot(x='Username', y='Likes', data=top_likes_users, color='green', label='Likes')

plt.title('Top 50 Users with the Highest Suscribers, Visits, and Likes')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```

```python
# Calculate average values
average_suscribers = df['Suscribers'].mean()
average_visits = df['Visits'].mean()
average_likes = df['Likes'].mean()
average_comments = df['Comments'].mean()

# Filter streamers with above-average performance
above_average_streamers = df[
    (df['Suscribers'] > average_suscribers) &
    (df['Visits'] > average_visits) &
    (df['Likes'] > average_likes) &
    (df['Comments'] > average_comments)
]


# Display the average values
print("Average Suscribers:", average_suscribers)
print("Average Visits:", average_visits)
print("Average Likes:", average_likes)
print("Average Comments:", average_comments)
print()

# Display the top-performing content creators
top_performers = above_average_streamers.sort_values(by='Rank').head(40)

print("Top-performing content creators:")
print(top_performers[['Rank', 'Username', 'Suscribers', 'Visits', 'Likes', 'Comments']])
```
